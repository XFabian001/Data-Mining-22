```python
import pandas as pd
from pandas import DataFrame
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

bach = pd.read_csv('https://raw.githubusercontent.com/zacharski/ml-class/master/data/bach.zip')

#Onehot encoding of feature data
notes = ('C',	'C#',	'D',	'D#',	'E',	'F',	'F#',	'G',	'G#', 'A', 'A#', 'B')

for note in notes:
  one_hot = pd.get_dummies(bach[[note]])
  bach = bach.drop(note, axis=1)
  bach = bach.join(one_hot)

one_hot = pd.get_dummies(bach['bass'])
bach = bach.drop('bass', axis=1)
one_hot = one_hot.add_prefix('bass_')
bach = bach.join(one_hot)

one_hot = pd.get_dummies(bach['meter'])
bach = bach.drop('meter', axis=1)
bach = bach.join(one_hot)

#One hot encoding the Choral IDs
one_hot = pd.get_dummies(bach['choral_ID'])
bach = bach.drop('choral_ID', axis=1)
one_hot = one_hot.add_prefix('choral_')
bach = bach.join(one_hot)

#bach = bach.drop('choral_ID', axis=1)

bach = bach.drop('event_number', axis=1)

#Function taken from pandas.pydata.org

bach = bach.drop_duplicates(keep="first", inplace=False)
```

```python
#Seperate data (random_state=0)
bfeatures = bach.drop('chord_label', axis=1)
blabels = bach[['chord_label']]
bf_train, bf_test, bl_train, bl_test = train_test_split(bfeatures, blabels, test_size = 0.2, random_state=0)

#CLF method (Default depth)
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(bf_train, bl_train)
bach_predict = clf.predict(bf_test)
score = accuracy_score(bl_test, bach_predict)
print(score)
```

```python
#Seperate data (random_state=42)
bfeatures = bach.drop('chord_label', axis=1)
blabels = bach[['chord_label']]
bf_train, bf_test, bl_train, bl_test = train_test_split(bfeatures, blabels, test_size = 0.2, random_state=42)

#CLF method (Default depth)
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(bf_train, bl_train)
bach_predict = clf.predict(bf_test)
score = accuracy_score(bl_test, bach_predict)
print(score)
```

Using an uncapped depth and random_state equal to zero, we were able to achieve the best results with 75.2% accuracy. The accuracy score dropped to 71.8% when the random_state was changed to 42, most likely due to the way the random number generator split the data. In an attempt to find out what would happen if I removed duplicate data values and one-hot encoded Choral ID values, I found that the results had dropped to 60% and 58%, respectively, after removing duplicate data values.
